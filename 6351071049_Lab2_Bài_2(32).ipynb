```python

import numpy as np

cm = np.array([
    [1146, 117],
    [2, 1173]
])

TP = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print("Trần Khôi Nguyên-6351071049")
print("=== TÍNH F-MEASURE TỪ CONFUSION MATRIX ===")
print("Precision =", round(precision, 4))
print("Recall    =", round(recall, 4))
print("F1-score  =", round(f1, 4))
```
